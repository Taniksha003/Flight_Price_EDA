```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
## Feature engineering
df['Date']= df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
#change datatype
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()
```

```python
#drop date of journey column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Extract arrival time without date
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
#split arrival time column into hours and minutes
df['Arrival_Hour']= df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min']=df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
#drop Arrival time 
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

```python
#change type of new columns
df['Arrival_Hour']= df['Arrival_Hour'].astype(int)
df['Arrival_Min']=df['Arrival_Min'].astype(int)
df.info()
```

```python
#Split dep_time column into Hour and minute
df['Dep_Hour']= df['Dep_Time'].str.split(':').str[0]
df['Dep_Min']=df['Dep_Time'].str.split(':').str[1]
```

```python
#change type of new columns
df['Dep_Hour']= df['Dep_Hour'].astype(int)
df['Dep_Min']=df['Dep_Min'].astype(int)
df.info()
```

```python
#Drop Dep_time column
df.drop('Dep_Time',axis=1,inplace=True)
df.head()
```

```python
#find unique values from total stops column
df['Total_Stops'].unique()
```

```python
#convert nan value into 1 stop
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1,'4 stops':4})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
#drop route column
df.drop('Route',axis=1,inplace=True)
df.head(2)
```

```python
#split Duration column into Hour and Min 
df['duration_Hour']= df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['duration_Min']= df['Duration'].str.split('h').str[1].str.split('m').str[0]
df.head()
```

```python
#find null values in duration_min column
df[df['duration_Min'].isnull()]
```

```python
#find unique values from duration_Min column
df['duration_Min'].unique()
```

```python
#find unique values from Airline column
df['Airline'].unique()
```

```python
#find unique values from Source column
df['Source'].unique()
```

```python
#find unique values from destination column
df['Destination'].unique()
```
